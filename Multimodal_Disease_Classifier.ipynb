```python
def categorize_problems(problems_str):
    if pd.isna(problems_str) or problems_str.strip() == '':
        return 'Normal'

    # Normalize the problems string to a list of individual problems
    problems = [label.strip().split('/')[0].strip() for label in problems_str.split(';') if label.strip()]

    has_cardiomegaly = 'Cardiomegaly' in problems
    has_pulmonary = any('Pulmonary' in p for p in problems)

    if has_cardiomegaly and has_pulmonary:
        return 'Cardiomegaly and Pulmonary'
    elif has_cardiomegaly:
        return 'Cardiomegaly'
    elif has_pulmonary:
        return 'Pulmonary'
    elif problems == ['normal']:
        return 'Normal'
    else:
        return 'Other'

merged_data['categorized_problem'] = merged_data['Problems'].apply(categorize_problems)

print("Value counts for the new 'categorized_problem' column:")
display(merged_data['categorized_problem'].value_counts())
```

```python
from transformers import TFAutoModel, AutoTokenizer
import tensorflow as tf
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.layers import Input, Dense, GlobalAveragePooling2D, concatenate, Dropout
from tensorflow.keras.models import Model

# Define constants for the text model (as they were implicitly used before)
MODEL_NAME = 'emilyalsentzer/Bio_ClinicalBERT'
MAX_SEQ_LEN = 128 # A common sequence length for BioClinicalBERT

# Instantiate tokenizer
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

# Tokenize text data for each split
train_encodings = tokenizer(train_df['combined_text_features'].tolist(), truncation=True, padding='max_length', max_length=MAX_SEQ_LEN, return_tensors='tf')
val_encodings = tokenizer(val_df['combined_text_features'].tolist(), truncation=True, padding='max_length', max_length=MAX_SEQ_LEN, return_tensors='tf')
test_encodings = tokenizer(test_df['combined_text_features'].tolist(), truncation=True, padding='max_length', max_length=MAX_SEQ_LEN, return_tensors='tf')

train_text_input_ids = train_encodings['input_ids']
train_attention_mask = train_encodings['attention_mask']

val_text_input_ids = val_encodings['input_ids']
val_attention_mask = val_encodings['attention_mask']

test_text_input_ids = test_encodings['input_ids']
test_attention_mask = test_encodings['attention_mask']

print("Text data tokenized for training, validation, and test sets.")

# Modified function to load and preprocess an image, returning only the image tensor
def load_and_preprocess_image_only(image_path):
    # Read the image file
    img = tf.io.read_file(image_path)
    # Decode the JPEG image
    img = tf.image.decode_jpeg(img, channels=3) # ensure 3 channels for RGB
    # Convert to float32 tensor
    img = tf.image.convert_image_dtype(img, tf.float32)
    # Resize the image
    img = tf.image.resize(img, [IMG_HEIGHT, IMG_WIDTH])
    # Apply DenseNet-specific preprocessing
    img = densenet.preprocess_input(img * 255) # preprocess_input expects 0-255 range
    return img

# Create image datasets for training, validation, and test sets
train_images_ds = tf.data.Dataset.from_tensor_slices(train_df['image_path'].tolist())
train_images_ds = train_images_ds.map(load_and_preprocess_image_only, num_parallel_calls=tf.data.AUTOTUNE)

val_images_ds = tf.data.Dataset.from_tensor_slices(val_df['image_path'].tolist())
val_images_ds = val_images_ds.map(load_and_preprocess_image_only, num_parallel_calls=tf.data.AUTOTUNE)

test_images_ds = tf.data.Dataset.from_tensor_slices(test_df['image_path'].tolist())
test_images_ds = test_images_ds.map(load_and_preprocess_image_only, num_parallel_calls=tf.data.AUTOTUNE)

# Create text input datasets
train_text_ds = tf.data.Dataset.from_tensor_slices({
    'input_ids': train_text_input_ids,
    'attention_mask': train_attention_mask
})
val_text_ds = tf.data.Dataset.from_tensor_slices({
    'input_ids': val_text_input_ids,
    'attention_mask': val_attention_mask
})
test_text_ds = tf.data.Dataset.from_tensor_slices({
    'input_ids': test_text_input_ids,
    'attention_mask': test_attention_mask
})

# Create label datasets
train_labels_ds = tf.data.Dataset.from_tensor_slices(train_df['encoded_labels'].tolist())
val_labels_ds = tf.data.Dataset.from_tensor_slices(val_df['encoded_labels'].tolist())
test_labels_ds = tf.data.Dataset.from_tensor_slices(test_df['encoded_labels'].tolist())

# Combine image, text inputs, and labels into a single dataset for each split
def combine_inputs_and_labels(image, text_inputs, label):
    text_inputs['image_input'] = image  # Add image to the text_inputs dictionary
    return text_inputs, label

train_dataset = tf.data.Dataset.zip((train_images_ds, train_text_ds, train_labels_ds))
train_dataset = train_dataset.map(combine_inputs_and_labels, num_parallel_calls=tf.data.AUTOTUNE)
train_dataset = train_dataset.shuffle(buffer_size=1024).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

val_dataset = tf.data.Dataset.zip((val_images_ds, val_text_ds, val_labels_ds))
val_dataset = val_dataset.map(combine_inputs_and_labels, num_parallel_calls=tf.data.AUTOTUNE)
val_dataset = val_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

test_dataset = tf.data.Dataset.zip((test_images_ds, test_text_ds, test_labels_ds))
test_dataset = test_dataset.map(combine_inputs_and_labels, num_parallel_calls=tf.data.AUTOTUNE)
test_dataset = test_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

print("Combined data pipelines created for training, validation, and test sets.")

# Verify the data pipeline
for inputs, labels in train_dataset.take(1):
    print("\nShape of image input in a training batch:", inputs['image_input'].shape)
    print("Shape of text input_ids in a training batch:", inputs['input_ids'].shape)
    print("Shape of text attention_mask in a training batch:", inputs['attention_mask'].shape)
    print("Shape of labels in a training batch:", labels.shape)

# Number of classes for classification
NUM_CLASSES = len(label_encoder.classes_)

# Define a custom Keras Layer to wrap the TFBertModel, now trainable
class FineTunableBertEncoderLayer(tf.keras.layers.Layer):
    def __init__(self, model_name, max_seq_len, **kwargs):
        super(FineTunableBertEncoderLayer, self).__init__(**kwargs)
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.bert_model = TFAutoModel.from_pretrained(model_name)
        self.max_seq_len = max_seq_len
        # Set BERT layers to trainable for fine-tuning
        self.bert_model.trainable = True

    def call(self, inputs):
        input_ids = inputs['input_ids']
        attention_mask = inputs['attention_mask']
        bert_output = self.bert_model(input_ids=input_ids, attention_mask=attention_mask)
        return bert_output.last_hidden_state[:, 0, :]

# --- Image Branch (DenseNet-121) ---
# Input for images
image_input = Input(shape=(IMG_HEIGHT, IMG_WIDTH, 3), name='image_input')

# Load pretrained DenseNet121 model (without top classification layer)
# Set trainable=True to allow fine-tuning
densenet_model = DenseNet121(weights='imagenet', include_top=False, input_tensor=image_input)
densenet_model.trainable = True # Ensure DenseNet layers are trainable

# Add a GlobalAveragePooling2D layer to flatten the output of DenseNet
image_features = GlobalAveragePooling2D()(densenet_model.output)

# --- Text Branch (BioClinicalBERT) ---
# Input for text (input_ids and attention_mask)
input_ids = Input(shape=(MAX_SEQ_LEN,), dtype=tf.int32, name='input_ids')
attention_mask = Input(shape=(MAX_SEQ_LEN,), dtype=tf.int32, name='attention_mask')

# Use the new fine-tunable BertEncoderLayer
bert_encoder = FineTunableBertEncoderLayer(MODEL_NAME, MAX_SEQ_LEN, name='fine_tunable_bert_encoder_layer')
text_features = bert_encoder({'input_ids': input_ids, 'attention_mask': attention_mask})

# --- Concatenate and Classification Head ---
# Combine image and text features
combined_features = concatenate([image_features, text_features])

# Add a dense layer for classification
x = Dense(256, activation='relu')(combined_features)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(NUM_CLASSES, activation='softmax')(x)

# Create the multimodal model
model = Model(inputs=[image_input, input_ids, attention_mask], outputs=output)

# Compile the model with a potentially lower learning rate for fine-tuning
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5), # Reduced learning rate for fine-tuning
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Multimodal model architecture reconstructed with fine-tunable encoders.")
model.summary()

print("Starting fine-tuning of the multimodal model...")

history_fine_tune = model.fit(
    train_dataset,
    epochs=EPOCHS, # Use the same number of epochs, or adjust as needed for fine-tuning
    validation_data=val_dataset
)

print("Fine-tuning complete.")
```
